# **1.Import Pandas DataFrame**

DataFrame adalah struktur data 2 dimensi yang berbentuk tabular (memiliki baris dan kolom). Hampir semua data tidak hanya memiliki 1 kolom tapi kolom nya lebih dari 1 kolom, sehingga cocok menggunakan Pandas DataFrame untuk mengolahnya.

Penggunaan variabel dataframe pada Python biasanya dengan menggunakan syntaks: df.

```python
import pandas as pd
```

# **2. Mengolah DataFrame**

```python
# Panggil file (iris.csv). lalu tampilkann 5 baris awal dataset dengan function head()
df = pd.read_csv("iris.csv")
df.head()
```

```python
# Tampilkan tipe data dari kolom yang ada pada dataset menggunakan dtypes
df.dtypes
```

Tipe Data dari kolom yang ada di dataset:

1.   Kolom "id" memiliki tipedata = Integer (int64).
2.   Kolom "SepalLenghtCm" memiliki tipedata = Float (float64).
3.   Kolom "Species" memiliki tipedata = Object (object).




```python
# Hitung ukuran (jumlah baris dan kolom) dari dataset
df.info()
```

```python
# Tampilkan data untuk kolom "id" dan kolom "species" dalam bentuk dataframe
df[["Id","Species"]]
```

```python
# Tampilkan data baris indeks ke-0 sampai dengan indeks ke - 9 
df.iloc[:10]
```

```python
# Tampilkan data hanya kolom "Id" dan kolom "Species", dan yang ditampilkan adalah data indeks ke - 11
# sampai dengan indeks ke - 15 
df[["Id","Species"]].iloc[11:16]
```

```python
# Tampilkan data 8 baris pertama
df.head(8)
```

```python
# Tampilkan data 3 baris terakhir
df.tail(3)
```

```python
# Hitung nilai mean dari dataset
df.mean()
```

```python
# Hitung nilai mean untuk kolom PetalLengthCm
df["PetalLengthCm"].mean()
```

```python
# Cari nilai minimal untuk kolom SepalWidthCm
df["SepalWidthCm"].min()
```

```python
# Hitung frekuensi pada kolom Species dengan menggunakan metode value_counts()
df["Species"].value_counts()
```

```python
# Tampilkan perhitungan frekuensi pada kolom Species dengan menggunakan value_counts() dalam bentuk dataframe
dfValueCountsSpecies = df["Species"].value_counts().rename_axis('Species Value Counts').reset_index(name='Counts')
dfValueCountsSpecies
```

```python
# Hitung frekuensi pada kolom PetalLenghCm dengan menggunakan value_counts() dalam bentuk dataframe
dfValueCountsPetalLengthCm = df["PetalLengthCm"].value_counts().rename_axis('PetalLengthCm Value Counts').reset_index(name='Counts')
dfValueCountsPetalLengthCm
```
